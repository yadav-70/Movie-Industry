```python
#Import Libraries 
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
% matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)
```

```python
# importing The data from excel and reading the data.
df=pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
#Checking the missing data
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#checking the datatypes of the data columns
df.dtypes
```

```python
#Changing the data types budget & gross 
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df.head()
```

```python
#Create a correct Year colummns
df['YearCorrected']=df['released'].astype(str)[:4]
df.head()
```

```python
#Sorting the data ascending the data
df.sort_values(by=['gross'],inplace=True,ascending=False)
```

```python
df.head()
```

```python
pd.set_option('display.max_rows',50)
```

```python
df.drop_duplicates().head()
```

```python
# Budget high correlation
# Company high correlation
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget Vs Gross Earning')
plt.xlabel('Gross Earnings')
plt.ylabel('Film Budget')
plt.show()
```

```python
df.head()
```

```python
#Planning Budget  Vs Gross Using Seaborn
plt.figure(figsize=(15,10))
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"Red"},line_kws={"color":"darkblue"})
```

```python
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```

```python
#High Correlations Between Budget Vs Gross Earning
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming correlation_matrix is already defined
plt.figure(figsize=(14, 10))  # Increase the size of the graph
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```

```python
df.head()
```

```python
df_numerized=df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
plt.figure(figsize=(15,10))
correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric features')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```

```python
df_numerized.corr().head()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[sorted_pairs>0.5]
high_corr
```

```python
#Voted and budget have the highest correlation gross earning
#Company has a low correlation
```

```python
#Final visualization of the correlations matrix with the correlations highlighted
plt.figure(figsize=(15,10))
sns.heatmap(correlation_matrix,annot=True,linewidths=1,mask=np.triu(correlation_matrix))
plt.title('Final Visualization correlation matrix of numeric features')
plt.xlabel('Movie Features')
plt.ylabel('movie Features')
plt.show()
```
